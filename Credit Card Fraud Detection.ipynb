# CREDIT CARD FRAUD DETECTION

## 1. Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

## 2. loading the dataset to a Pandas DataFrame

```python
credit_card_data = pd.read_csv('creditcard.csv')
```

## 3. first 5 rows of the dataset

```python
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# dataset informations
credit_card_data.info()
```

## 4. checking the number of missing values in each column

```python
credit_card_data.isnull().sum()
```

## 5. distribution of legit transactions & fraudulent transactions

```python
credit_card_data['Class'].value_counts() #0 --> Normal Transaction
#1 --> fraudulent transaction
```

## 6.separating the data for analysis

```python
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

## 7. statistical measures of the data

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

## 8. compare the values for both transactions

```python
credit_card_data.groupby('Class').mean()
```

### Under-Sampling

Build a sample dataset containing similar distribution of normal transactions and Fraudulent Transactions

Number of Fraudulent Transactions --> 492

```python
legit_sample = legit.sample(n=492)
```

## 9. Concatenating two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

## 10. Splitting the data into Features & Targets

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

## 11. Split the data into Training data & Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Model Training

## Logistic Regression

```python
model = LogisticRegression()
```

## 12. training the Logistic Regression Model with Training Data

```python
model.fit(X_train, Y_train)
```

# Model Evaluation

## 13. Accuracy Score

```python
#accuracy on train data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
